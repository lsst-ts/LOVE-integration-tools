```python
from lsst.ts import salobj
```

```python
r = salobj.Remote(salobj.Domain(), 'ATSpectrograph')
await r.start_task
print("Remote connected...")

evt = await r.evt_summaryState.aget()
state = salobj.State(evt.summaryState)
print(state)

if(state == salobj.State.STANDBY):
    await r.cmd_start.start(timeout=30)
    await r.cmd_enable.start(timeout=30)

evt = await r.evt_summaryState.aget()
state = salobj.State(evt.summaryState)
print(state)

events = [d for d in dir(r) if 'evt' in d]
telemetries = [d for d in dir(r) if 'tel' in d]
print(events, telemetries)

for e in events:
    print(e)
    evt_remote = getattr(r, e)
    data = evt_remote.get()
    print(data)

for t in telemetries:
    print(t)
    tel_remote = getattr(r, t)
    data = tel_remote.get()
    print(data)
```
